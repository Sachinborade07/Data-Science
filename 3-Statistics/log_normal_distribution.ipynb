```python
import pandas as pd
import seaborn as sns
import numpy as np 
```

```python
df = pd.read_csv(
    "csv/income(1).csv",
    index_col = None,
    names = ['income','count'],
    skiprows=1
)
```

```python
df.head()
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
g = sns.barplot(x='income',y='count',data = df)
g.set_xticklabels(g.get_xticklabels(),horizontalalignment='right');
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
g = sns.barplot(x='income',y='count',data = df)
g.set_xticklabels(g.get_xticklabels(),rotation=45,horizontalalignment='right');
# now the result is not messy 
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
g = sns.barplot(x='income',y='count',data = df)
g.set_xticklabels(g.get_xticklabels(),rotation=45,horizontalalignment='right');
g.set(xscale='log')
```
